# Data visualisation de dépots open-source d'organisations publics

Source des données : https://www.data.gouv.fr/fr/datasets/inventaire-des-depots-de-code-source-des-organismes-publics/

Je cherche avec ce dataset à faire ressortir quels sont les languages et les plateformes les plus dévellopé dans le domaine de l'open source.

### 1. Imports des librairies d'extractions et d'affichage des données

```python
import pandas
import numpy
import matplotlib
import seaborn
```

### 2. Chargement des données 

On charge les données directement depuis le site pour ne pas avoir gérer un fichier csv à côté de notre notebook.

```python
raw_data = pandas.read_csv("https://raw.githubusercontent.com/etalab/data-codes-sources-fr/master/data/repertoires/csv/all.csv")
raw_data
```

### 3. Etude des diférentes plateformes

```python
platform_data = raw_data.groupby(["plateforme"]).count()
platform_data
```

```python
platform_data["nom"].plot.pie(autopct="%.1f%%")
```

On remarque que **GitHub** est une plateforme largement plus utilisé que **GitLab**.

### 4. Etude des différents languages

```python
lang_data = raw_data.groupby(["langage"]).count()["nom"].argsort()
lang_data
```

```python
lang_data.plot.pie(autopct="%.2f%%")
```
